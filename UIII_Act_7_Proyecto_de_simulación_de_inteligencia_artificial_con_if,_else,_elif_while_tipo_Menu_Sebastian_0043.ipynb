### Proyecto de simulación de inteligencia artificial con if, else, elif  while tipo Menu
### 18. Administrador y empleados

```python
print("--- IA Empresarial v1.0 ---")
estado = "inicio"
while estado == "inicio":
    print("\n¿Quién eres? (admin / empleado / salir)")
    usuario = input("Tu respuesta: ")

    if usuario == "admin":
        print("Bienvenido, Jefe. ¿Qué deseas hacer? (ver / despedir)")
        accion = input()
        if accion == "ver":
            print("Analizando datos... Todos están trabajando duro.")
        elif accion == "despedir":
            print("IA: Sugiero dar una segunda oportunidad.")
        else:
            print("Opción no válida.")

    elif usuario == "empleado":
        print("Hola. ¿Cómo te sientes hoy? (feliz / cansado)")
        sentimiento = input()
        if sentimiento == "feliz":
            print("¡Excelente! La productividad subirá un 20%.")
        else:
            print("IA: Te recomiendo tomar un café de 5 minutos.")

    elif usuario == "salir":
        estado = "fin"
        print("Cerrando sistema...")
    else:
        print("Usuario no reconocido.")
```
