### Proyecto de simulación de inteligencia artificial con if, else, elif  while tipo Menu
### 18. Administrador y empleados

```python
print("--- IA Empresarial v1.0 ---")
estado = "inicio"

while estado == "inicio":
    print("\n--- MENÚ PRINCIPAL ---")
    print("1. Administrador")
    print("2. Empleado")
    print("3. Salir")

    usuario = input("Selecciona una opción (1-3): ")

    if usuario == "1":
        print("\n--- BIENVENIDO, JEFE ---")
        print("1. Ver estado del equipo")
        print("2. Despedir a alguien")
        accion = input("Selecciona una acción (1-2): ")

        if accion == "1":
            print("IA: Analizando datos... Todos están trabajando duro.")
        elif accion == "2":
            print("IA: Sugiero dar una segunda oportunidad.")
        else:
            print("Opción no válida.")

    elif usuario == "2":
        print("\n--- HOLA, EMPLEADO ---")
        print("¿Cómo te sientes hoy?")
        print("1. Feliz")
        print("2. Cansado")
        sentimiento = input("Selecciona una opción (1-2): ")

        if sentimiento == "1":
            print("IA: ¡Excelente! La productividad subirá un 20%.")
        elif sentimiento == "2":
            print("IA: Te recomiendo tomar un café de 5 minutos.")
        else:
            print("Opción no válida.")

    elif usuario == "3":
        estado = "fin"
        print("Cerrando sistema...")

    else:
        print("Opción no reconocida. Por favor, introduce 1, 2 o 3.")
```
